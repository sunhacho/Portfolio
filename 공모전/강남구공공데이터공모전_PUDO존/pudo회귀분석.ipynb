## **다중회귀분석 (변수별 중요도 분석)**

```python
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error

# 데이터셋을 feature와 target으로 분리
X = df.drop(columns=['도로명주소_그룹', '불법주정차수'])  # feature
y = df['불법주정차수']  # target

# 데이터 스케일링
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 학습용과 평가용으로 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=1)

# 릿지 회귀 모델 학습 및 최적의 alpha 값 찾기
ridge = Ridge()
param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100,1000,4000]}
grid_search = GridSearchCV(ridge, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)
best_alpha = grid_search.best_params_['alpha']

# 최적의 alpha 값을 사용하여 모델 재학습
ridge_reg = Ridge(alpha=best_alpha)
ridge_reg.fit(X_train, y_train)

# 학습된 모델로 예측
y_pred_train = ridge_reg.predict(X_train)
y_pred_test = ridge_reg.predict(X_test)

# 평가 메트릭 계산 (RMSE)
train_rmse = mean_squared_error(y_train, y_pred_train, squared=False)
test_rmse = mean_squared_error(y_test, y_pred_test, squared=False)

print("Best alpha:", best_alpha)
print("Train RMSE:", train_rmse)
print("Test RMSE:", test_rmse)

# 회귀 계수 출력
coefficients = ridge_reg.coef_
print("회귀 계수:", coefficients)
```
