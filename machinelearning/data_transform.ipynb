```python
chat_data = pd.read_csv('/content/data1.csv',engine='python',encoding='utf-8')
chat_data
```

```python
del chat_data['Unnamed: 0']
del chat_data['Language']
```

```python
chat_data
```

```python
list_text = []

for i in trange(len(chat_data)):

    pre_text = chat_data['Text'][i]
        # Retweets Ï†úÍ±∞
    text = re.sub('RT @[\w_]+: ', '', pre_text)

        # enticons Ï†úÍ±∞
    text = re.sub('@[\w_]+', '', text)

        # URL Ï†úÍ±∞
    text = re.sub(r"http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+", ' ',
                      text)  # httpÎ°ú ÏãúÏûëÎêòÎäî url
    text = re.sub(r"[-a-zA-Z0-9@:%._\+~#=]{1,256}\.[a-zA-Z0-9()]{2,6}\b([-a-zA-Z0-9()@:%_\+.~#?&//=]*)", ' ',
                      text)  # httpÎ°ú ÏãúÏûëÎêòÏßÄ ÏïäÎäî url
        #     pattern = '(http|ftp|https)://(?:[-\w.]|(?:\da-fa-F]{2}))+'
        #     text = re.sub(pattern = pattern, repl = ' ',string=text)

        # Hashtag Ï†úÍ±∞
    text = re.sub('[#]+[0-9a-zA-Z_]+', ' ', text)

        # Ïì∞Î†àÍ∏∞ Îã®Ïñ¥ Ï†úÍ±∞
    text = re.sub('[&]+[a-z]+', ' ', text)

        # ÌäπÏàòÎ¨∏Ïûê Ï†úÍ±∞
    text = re.sub('[^0-9a-zA-Z„Ñ±-„ÖéÍ∞Ä-Ìû£]', ' ', text)

        # ÎùÑÏñ¥Ïì∞Í∏∞ Ï†úÍ±∞
    text = text.replace('\n', ' ')

    text = re.sub(r'can t', 'can not', text)
    text = re.sub(r'don t', 'do not', text)
    text = re.sub(r'couldn t', 'could not', text)

    #ÏÜåÎ¨∏Ïûê ÎåÄÎ¨∏Ïûê
    text = text.lower()

    # if Num is True:
    #         # Ïà´Ïûê Ï†úÍ±∞
    #     text = re.sub(r'\d+', ' ', text)

        # Ï†ïÎ¶¨
    text = ' '.join(text.split())
    list_text.append(text)
```

```python
real_df=pd.DataFrame(list_text, columns = ['text'])
real_df
```
